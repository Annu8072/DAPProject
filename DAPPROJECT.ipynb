```python
data = {
    'team': ['team1', 'team2'],
    'total_attempts': [35, 28], 
}
df = pd.DataFrame(data)

# Create the bar plot
sns.barplot(x='team', y='total_attempts', data=df, palette="viridis")

# Add title and labels
plt.title('Total Attempts by Team')
plt.xlabel('Teams')
plt.ylabel('Total Attempts')

# Show the plot
plt.show()
```

```python
data = {
    'Metric': ['Total Attempts', 'Passes', 'Goals Scored', 'Corners'],
    'Team1': [35, 420, 3, 8],  # Replace with values for team1
    'Team2': [28, 390, 2, 6],  # Replace with values for team2
}

df = pd.DataFrame(data)
df_melted = df.melt(id_vars='Metric', var_name='Team', value_name='Value')
sns.barplot(x='Metric', y='Value', hue='Team', data=df_melted, palette='viridis')
plt.title('Comparison of Metrics Between Team1 and Team2')
plt.xlabel('Metrics')
plt.ylabel('Values')
plt.xticks(rotation=30)
plt.show()
```

```python

data = {
    'Team': ['Team1', 'Team2'],
    'Passes': [420, 390],  # Replace these with actual values
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Create the pie chart
plt.figure(figsize=(6, 6))
plt.pie(
    df['Passes'], 
    labels=df['Team'], 
    autopct='%1.1f%%',  # Show percentages
    startangle=90,  # Align start angle
    colors=['#FFD700', '#FF4500'],  # Custom colors
    explode=(0.05, 0),  # Slightly separate Team1
    shadow=True  # Add shadow for aesthetics
)

# Add a title
plt.title('Passes Comparison Between Team1 and Team2')

# Show the plot
plt.show()
```

```python
data = {
    'total_attempts_team1': [5, 8, 7, 10, 15, 12, 11, 14, 9, 6],
    'total_attempts_team2': [6, 9, 8, 11, 13, 15, 10, 12, 8, 7]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Create a histogram for total_attempts_team1
plt.figure(figsize=(8, 6))
sns.histplot(df['total_attempts_team1'], bins=8, kde=True, color='blue', label='Team1 Attempts')

# Add title and labels
plt.title('Distribution of Total Attempts for Team1')
plt.xlabel('Total Attempts')
plt.ylabel('Frequency')
plt.legend()

# Show the plot
plt.show()
```

```python
# Create a histogram comparing total attempts for both teams
plt.figure(figsize=(8, 6))
sns.histplot(df['total_attempts_team1'], bins=8, kde=True, color='blue', label='Team1 Attempts', alpha=0.6)
sns.histplot(df['total_attempts_team2'], bins=8, kde=True, color='orange', label='Team2 Attempts', alpha=0.6)

# Add title and labels
plt.title('Comparison of Total Attempts for Team1 and Team2')
plt.xlabel('Total Attempts')
plt.ylabel('Frequency')
plt.legend()

# Show the plot
plt.show()
```
